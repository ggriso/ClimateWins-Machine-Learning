## 3. Scaling Data

```python
# Create scaler object
# StandardScaler assumes data is normally distributed and scales with a distribution around 0
# and standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
```

```python
# Create a new df with scaled data

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns = list(df))

df_scaled.head()
```

```python
# Compare the original vs. scaled data

df.head()
```

#### We should remove scaled Date and Month columns from the scaled dataset (as they represent dates, not numbers), and then attach the original Date and Month columns to prevent loss of information.

```python
# Create a copy of the DATE and MONTH columns

date_month_columns = df[['DATE', 'MONTH']]
```

```python
# Drop unnecessary scaled columns

df_scaled.drop(['DATE', 'MONTH'], axis=1, inplace=True)
```

```python
df_scaled.head()
```

```python
# Concatenate the DATE and MONTH columns back to the scaled dataframe

df_scaled_final = pd.concat([date_month_columns, df_scaled], axis=1)
```

```python
df_scaled_final
```

## 4. Exporting Scaled Dataframe

```python
df_scaled_final.to_csv(os.path.join(path, 'dataset_scaled.csv'), index=False)
```
